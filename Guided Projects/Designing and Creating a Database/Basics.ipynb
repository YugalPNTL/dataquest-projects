## Importing Data into SQLite

### Create Helper Functions

```python
# These helper functions will be useful as we work
# with the SQLite database from python

DB = 'mlb.db'

def run_query(q):
    # Take an SQL query and return a pandas dataframe of the query
    with sqlite3.connect(DB) as conn:
        return pd.read_sql(q, conn)
    
def run_command(c):
    # Take an SQL command and execute it
    with sqlite3.connect(DB) as conn:
        conn.execute('PRAGMA foreign_keys = ON;')
        conn.isolation_level = None
        conn.execute(c)
    
def show_tables():
    # Return a list of all tables and views in the database
    q = '''SELECT
                name,
                type
            FROM sqlite_master
            WHERE type IN ("table","view");'''
    return run_query(q)
```

### Create Tables and Add game_id 

```python
with sqlite3.connect('mlb.db') as conn:
    game_log.to_sql('game_log', conn, index = False, if_exists = 'replace')
    park_codes.to_sql('park_codes', conn, index = False, if_exists = 'replace')
    person_codes.to_sql('person_codes', conn, index = False, if_exists = 'replace')
    team_codes.to_sql('team_codes', conn, index = False, if_exists = 'replace')
    
show_tables()
```

## Creating Tables Without Foreign Key Relations

#### Person Table

```python
create_person = """
    CREATE TABLE IF NOT EXISTS person(
        person_id TEXT PRIMARY KEY,
        first_name TEXT,
        last_name TEXT);"""

# populate person with fields from person_codes
populate_person = """
    INSERT OR IGNORE INTO person
    SELECT 
        id,
        first,
        last
    FROM person_codes;"""

run_command(create_person)
run_command(populate_person)

run_query("SELECT * FROM person LIMIT 5;")
```

#### League Table

```python
create_league = """
    CREATE TABLE IF NOT EXISTS league(
        league_id TEXT PRIMARY KEY,
        name TEXT);"""

# populate league with researched league names from above
populate_league = """
    INSERT OR IGNORE INTO league
    VALUES ('NL','National League'),
        ('AL','American League'),
        ('AA','American Association'),
        ('FL','Federal League'),
        ('PL',"Players' League"),
        ('UA','Union Association');"""

run_command(create_league)
run_command(populate_league)

run_query("SELECT * FROM league")
```

#### Appearance Type Table

##  Adding the Team and Game Tables

Next we add the team and game tables because they need to exist before we create the appearance tables that will point to them.

#### Team Table

```python
create_team = """
    CREATE TABLE IF NOT EXISTS team(
        team_id TEXT PRIMARY KEY,
        league_id TEXT,
        city TEXT,
        nickname TEXT,
        franch_id TEXT,
        FOREIGN KEY (league_id) REFERENCES league(league_id));"""

# populate team with fields from team_codes
populate_team = """
    INSERT OR IGNORE INTO team
    SELECT 
        team_id,
        league,
        city,
        nickname,
        franch_id
    FROM team_codes;"""

run_command(create_team)
run_command(populate_team)

run_query("SELECT * FROM team LIMIT 5;")
```

#### Game Table

```python
create_game = """
    CREATE TABLE IF NOT EXISTS game(
        game_id TEXT PRIMARY KEY,
        date DATE,
        number_of_game INTEGER,
        park_id TEXT,
        length_outs INTEGER,
        day BOOLEAN,
        completion TEXT,
        forfeit TEXT,
        protest TEXT,
        attendance INTEGER,
        length_minutes INTEGER,
        additional_info TEXT,
        acquisition_info TEXT,
        FOREIGN KEY (park_id) REFERENCES park(park_id));"""


# populate game with game-specific fields from game_log
populate_game = """
    INSERT OR IGNORE INTO game
    SELECT game_id,
        date,
        number_of_game,
        park_id,
        length_outs,
        CASE 
            WHEN day_night = 'D' THEN 1
            WHEN day_night = 'N' THEN 0
            ELSE NULL
            END as day,
        completion,
        forefeit,
        protest,
        attendance,
        length_minutes,
        additional_info,
        acquisition_info
    FROM game_log;"""

run_command(create_game)
run_command(populate_game)

run_query("SELECT * FROM game LIMIT 5;")
```

## Adding the Team Appearance Table

Start by getting the schema from game_log in order create and populate the new table without having to type every single column.

```python
game_log_schema = run_query("""
                SELECT sql FROM sqlite_master
                WHERE name = "game_log"
                  AND type = "table";""").iloc[0,0]
# Split the schema out into the strings used to create individual fields
game_log_fields = game_log_schema.split('\n')[1:]
game_log_fields = [field.strip().replace('"','') for field in game_log_fields]
game_log_fields
```

For creating the team_appearance table, use the stats fields which are elements 21 through 48 in the list of fields from the game_log schema. We'll remove the prefix 'v_' to make these new fields universal and change all of the stats fields to integer as this will not lose any information and will be more convenient.

```python
# Make writing the create command easier by identifying the stats fields,
# removing the 'v_' prefix and creating a single string out of them.
create_fields = [field.replace('v_','').replace('REAL','INTEGER') for field in game_log_fields]
create_fields = create_fields[21:49]
create_fields_string = '\n        '.join(create_fields)

create_team_appearance = """
    CREATE TABLE IF NOT EXISTS team_appearance(
        team_id TEXT,
        game_id TEXT,
        home BOOLEAN,
        league_id TEXT,
        score INTEGER,
        line_score TEXT,
        """ + create_fields_string + """
        PRIMARY KEY (team_id, game_id),
        FOREIGN KEY (team_id) REFERENCES team(team_id),
        FOREIGN KEY (game_id) REFERENCES game(game_id),
        FOREIGN KEY (league_id) REFERENCES league(league_id));"""
print(create_team_appearance)
```

For populating the table create strings of home and visiting stats columns to be used in writing the command. As noted before the visiting team stats are in fields 21 through 48. The home team stats are in fields 49 through 76.

```python
# Need to strip out the type from each field, as these are
# going to be used in select statements. This isn't too difficult
# since each fields is of type REAL.
visiting_stats_fields = [field.replace(' REAL','') for field in game_log_fields[21:49]]
home_stats_fields = [field.replace(' REAL','') for field in game_log_fields[49:77]]

# create strings, removing last comma as these are the last fields in the select
visiting_stats_string = '\n        '.join(visiting_stats_fields)[:-1]
home_stats_string = '\n        '.join(home_stats_fields)[:-1]

populate_team_appearance = """
    INSERT OR IGNORE INTO team_appearance
    SELECT
        h_name,
        game_id,
        1,
        h_league,
        h_score,
        h_line_score,
        """ + home_stats_string + """
    FROM game_log
    
    UNION
    
    SELECT
        v_name,
        game_id,
        0,
        v_league,
        v_score,
        v_line_score,
        """ + visiting_stats_string + """
    FROM game_log;"""

#print(populate_team_appearance)
```

With the commands written - create, populate, and verify the new table.

```python
run_command(create_team_appearance)
run_command(populate_team_appearance)

run_query("""
    SELECT * FROM team_appearance
    WHERE game_id = (SELECT game_id FROM game WHERE date = (SELECT MIN(date) FROM game)) OR game_id = (SELECT MAX(game_id) FROM game WHERE date = (SELECT MAX(date) FROM game))
    ORDER BY game_id, home;""")
```

## Adding the Person Appearance Table

First a query to populate the table with all data from game_log columns for managers, umpires, pitchers and awards. This is quite long as we need to write a seperate SELECT statement for each of these columns.

```python
create_person_appearance = """
    CREATE TABLE IF NOT EXISTS person_appearance(
        appearance_id INTEGER PRIMARY KEY,
        person_id TEXT,
        team_id TEXT,
        game_id TEXT,
        appearance_type_id TEXT,
        FOREIGN KEY (person_id) REFERENCES person(person_id),
        FOREIGN KEY (team_id) REFERENCES team(team_id),
        FOREIGN KEY (game_id) REFERENCES game(game_id),
        FOREIGN KEY (appearance_type_id) REFERENCES appearance_type(appearance_type_id));"""

populate_person_appearance_nonplayers = """
    INSERT OR IGNORE INTO person_appearance (
        game_id,
        team_id,
        person_id,
        appearance_type_id
    )
        SELECT
            game_id,
            NULL,
            hp_umpire_id,
            "UHP"
        FROM game_log
        WHERE hp_umpire_id IS NOT NULL

    UNION
    
        SELECT
            game_id,
            NULL,
            [1b_umpire_id],
            "U1B"
        FROM game_log
        WHERE [1b_umpire_id] IS NOT NULL

    UNION
    
        SELECT
            game_id,
            NULL,
            [2b_umpire_id],
            "U2B"
        FROM game_log
        WHERE [2b_umpire_id] IS NOT NULL

    UNION
    
        SELECT
            game_id,
            NULL,
            [3b_umpire_id],
            "U3B"
        FROM game_log
        WHERE [3b_umpire_id] IS NOT NULL

    UNION
    
        SELECT
            game_id,
            NULL,
            lf_umpire_id,
            "ULF"
        FROM game_log
        WHERE lf_umpire_id IS NOT NULL

    UNION

        SELECT
            game_id,
            NULL,
            rf_umpire_id,
            "URF"
        FROM game_log
        WHERE rf_umpire_id IS NOT NULL

    UNION

        SELECT
            game_id,
            v_name,
            v_manager_id,
            "MM"
        FROM game_log
        WHERE v_manager_id IS NOT NULL

    UNION

        SELECT
            game_id,
            h_name,
            h_manager_id,
            "MM"
        FROM game_log
        WHERE h_manager_id IS NOT NULL

    UNION

        SELECT
            game_id,
            CASE
                WHEN h_score > v_score THEN h_name
                ELSE v_name
                END,
            winning_pitcher_id,
            "AWP"
        FROM game_log
        WHERE winning_pitcher_id IS NOT NULL

    UNION

        SELECT
            game_id,
            CASE
                WHEN h_score > v_score THEN v_name
                ELSE h_name
                END,
            losing_pitcher_id,
            "ALP"
        FROM game_log
        WHERE losing_pitcher_id IS NOT NULL
        
    UNION

        SELECT
            game_id,
            CASE
                WHEN h_score > v_score THEN h_name
                ELSE v_name
                END,
            saving_pitcher_id,
            "ASP"
        FROM game_log
        WHERE saving_pitcher_id IS NOT NULL
        
    UNION

        SELECT
            game_id,
            CASE
                WHEN h_score > v_score THEN h_name
                ELSE v_name
                END,
            winning_rbi_batter_id,
            "AWB"
        FROM game_log
        WHERE winning_rbi_batter_id IS NOT NULL
        
    UNION
        
        SELECT
            game_id,
            h_name,
            h_starting_pitcher_id,
            "PSP"
        FROM game_log
        WHERE h_starting_pitcher_id IS NOT NULL
        
    UNION
    
        SELECT
            game_id,
            v_name,
            v_starting_pitcher_id,
            "PSP"
        FROM game_log
        WHERE v_starting_pitcher_id IS NOT NULL"""
```

To avoid writing an even longer query to populate the table with offensive and defensive player appearances - loop through all permutations of home/away and player number 1-10 and add each players offensive and defensive appearance to the new table using python string formatting.

```python
populate_player_template = """
    INSERT INTO person_appearance (
        game_id,
        team_id,
        person_id,
        appearance_type_id
    ) 
        SELECT
            game_id,
            {hv}_name,
            {hv}_player_{num}_id,
            "O{num}"
        FROM game_log
        WHERE {hv}_player_{num}_id IS NOT NULL

    UNION

        SELECT
            game_id,
            {hv}_name,
            {hv}_player_{num}_id,
            "D" || CAST({hv}_player_{num}_def_pos AS INT)
        FROM game_log
        WHERE {hv}_player_{num}_id IS NOT NULL;
    """

run_command(create_person_appearance)
run_command(populate_person_appearance_nonplayers)

for hv in ["h","v"]:
    for num in range(1,10):
        query_vars = {
            "hv": hv,
            "num": num
        }
        run_command(populate_player_template.format(**query_vars))
        
```
