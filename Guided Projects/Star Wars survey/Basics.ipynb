```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.head(10)
```

```python
yes_no = {"Yes": True, "No": False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

```python
import numpy

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    numpy.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
    

star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns = {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': "ranking_1",
    'Unnamed: 10': "ranking_2",
    'Unnamed: 11': "ranking_3",
    'Unnamed: 12': "ranking_4",
    'Unnamed: 13': "ranking_5",
    'Unnamed: 14': "ranking_6"    
})
star_wars.columns
```

```python
%matplotlib inline
average_ranking = star_wars[star_wars.columns[9:15]].mean()
average_ranking.plot.bar()
```

So far, we've cleaned up the data, renamed several columns, and computed the average ranking of each movie. It looks like the original trilogy is rated better than the newer three star wars movies.

```python
total_seen = star_wars[star_wars.columns[3:9]].sum()
total_seen.plot.bar()
```

In general, it looks like the original trilogy has been seen more than the prequels. However, more people have seen Episode 1 than the original Star Wars movie (Episode 4). So in genearl, the movies ranked better have been seen more.

```python
fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == True]
nonfans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == False]
```

```python
fans[fans.columns[9:15]].mean().plot.bar()
```

```python
nonfans[nonfans.columns[9:15]].mean().plot.bar()
```

Non-fans rank the prequels much better than fans compared to the original trilogy. Fans hate the prequels.

```python
fans[fans.columns[3:9]].sum().plot.bar()
```

```python
nonfans[nonfans.columns[3:9]].sum().plot.bar()
```

However, fans are more likely to have seen the prequels than non-fans. Also, Episode 4 is not popular among nonfans.
