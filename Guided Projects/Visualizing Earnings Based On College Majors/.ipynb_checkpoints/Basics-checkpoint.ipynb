```python
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
recent_grads.head()
recent_grads.tail()
```

```python
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print(raw_data_count,'/',cleaned_data_count)
```

# Pandas, Scatter Plots

```python
ax = recent_grads.plot(kind='scatter', x="Sample_size" ,y="Median",title = "Median vs. Sample Size")
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment Rate vs. Sample Size')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full Time')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment Rate vs. Share Women')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs. Number Men')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs. Number of Women')
```

No strong correlations are noted among the relations plotted.

# Pandas, Scatter Matrix Plot

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(recent_grads[["Sample_size","Median"]], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[["Sample_size","Median","Unemployment_rate"]], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[["ShareWomen","Median","Unemployment_rate"]])
```

In this last plot we can see a downward trend for median salary as the share of women in the major gets higher. No correlation is seen between share of women and unemployment rate.

# Pandas, Bar Plots

```python
recent_grads[:10].plot.bar(x='Major',y='ShareWomen',ylim=(0,1),legend=False)
recent_grads[-10:].plot.bar(x='Major',y="ShareWomen",ylim=(0,1),legend=False)
```

```python
recent_grads[:10].plot.bar(x='Major',y='Unemployment_rate',ylim=(0,.2),legend=False)
recent_grads[-10:].plot.bar(x='Major',y="Unemployment_rate",ylim=(0,.2),legend=False)
```

# Pandas, Grouped Bar Plot

```python
recent_grads[['Major_category','Men','Women']].groupby(by='Major_category').agg('sum').plot.bar()
```

```python
group_table = recent_grads.pivot_table(index='Major_category', values=['Women','Men'], aggfunc=sum) 
group_table.plot.bar()
```
